### **Задание 11**
Здесь нужно знать, что с помощью $K$ бит мы можем закодировать $2^K$ различных вариантов. Также в этом задании принято за **ИСТИНУ СВЫШЕ**, что в $1 байт = 8 бит$

Нельзя выделить общий алгоритм, так как всё зависит от условия задания. Можно выделить что-то типа алгоритма:
0. Внимательно читаем и понимаем, что от нас хотят
1. Понимаем сколько символов в нашем алфавите
2. Считаем необходимое кол-во битов, которыми мы кодируем каждый символ (если надо)
3. Считаем необходимое кол-во байт, которыми нужно закодировать наш пароль (если надо)
4. Проводим финальные вычисления
5. Проверяем ход решения и пишем ответ


**ПРИМЕР:**

Автомобильный номер состоит из 6 символов. Допустимыми символами считаются 10 цифр и 5 заглавных букв: A, P, T, E, K. Для хранения каждого из 15 допустимых символов используется одинаковое и наименьшее возможное количество бит. Для хранения каждого номера используется одинаковое и минимально возможное количество байт. Сколько байт памяти потребуется для хранения 400 автомобильных номеров? Номера хранятся без разделителей.

0. Прочитали
1. В алфавите $15$ символов - $10$ цифр и $5$ заглавных букв
2. Тогда нужно решить уравнение $2^x > 15$, Отсюда $x=4$, т.е. нам надо по $4$ бита на каждый символ, чтобы мы закодировали каждый символ одинаковым и минимиально возможным количеством бит.
3. Наш пароль состоит из $6$ символов, каждый весит по $4$ бит, тогда весь пароль весит $6*4=24$ бита. Тогда посчитаем сколько минимум байт нам потребуется: $ {\displaystyle \lceil {\frac{24}{8}}\rceil = 3 байт}$
4. Это мы посчитали сколько нужно для ОДНОГО номера. Тогда для 400 номеров потребуется: $3*400=1200 байт$
5. Пишем ответ
